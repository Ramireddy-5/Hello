```python
import nltk
from nltk.corpus import genesis
from nltk.probability import ConditionalFreqDist
from nltk.util import bigrams

# Download required resources if not already present
nltk.download("genesis")
nltk.download("punkt")

# Load the corpus
words = genesis.words("english-web.txt")

# Create bigrams from the corpus
bigram_list = list(bigrams(words))

# Build a Conditional Frequency Distribution from the bigrams
cfd = ConditionalFreqDist(bigram_list)

# Starting word
start_word = "darkness"

# Generate a 25-word sentence
sentence = [start_word]
for i in range(24):  # Already have 1 word, so add 24 more
    current_word = sentence[-1]
    if current_word in cfd:
        # Pick the most likely next word
        next_word = cfd[current_word].max()
    else:
        # If no continuation, stop early
        break
    sentence.append(next_word)

# Join the words to form a sentence
print("Generated Sentence:\n")
print(" ".join(sentence))
```
